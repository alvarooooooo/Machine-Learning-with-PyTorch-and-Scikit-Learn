# 1. Giving Computers the Ability to Learn from Data

* **Supervised Learning:** Labeled data, Direct feedback and Predict outcome/future
* **Unsupervised Learning:** No labels/targets, No feedback and Find hidden structure in data
* **Reinforcement Learning:** Decision process, Reward system and Learn series of actions

### NO FREE LUNCH THEOREM
* Author: David Wolpert 1996
* There are no easy shortcuts to success. No single algorithm is universally better than all others for every problem.
